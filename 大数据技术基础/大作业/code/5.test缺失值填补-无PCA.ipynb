```python
import pandas as pd
import numpy as np
```

```python
all_features=pd.read_csv("test_all_features_noPCA.csv")
all_features.head()
```

```python
# 用K近邻算法填补缺失值
from sklearn.impute import KNNImputer
clf=KNNImputer()
all_features_withoutnan=clf.fit_transform(all_features)
all_features_withoutnan.shape
```

```python
# 
if all_features_withoutnan.shape==all_features.shape:
    print("最初all_features的缺失值个数：",np.sum(np.sum(all_features.isna())))
    all_features=pd.DataFrame(all_features_withoutnan,columns=all_features.columns)
    print("已经把填补过的值放进去")
    print("现在all_features的缺失值个数：",np.sum(np.sum(all_features.isna())))
else:
    raise BaseException("出事了！")
```

```python
# 首先对数据进行归一化
from sklearn import preprocessing
all_features.iloc[:,4:]=preprocessing.normalize(all_features.iloc[:,4:])
all_features.head()
```

```python
all_features.to_csv("test_all_features_noPCA-2.csv")
```
